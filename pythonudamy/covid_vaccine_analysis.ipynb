# month and vaccination

```python
months = range(1,13)
plt.bar(months,df.groupby(['month']).sum()['Total Individuals Vaccinated'])

plt.xticks(months)
plt.ylabel('vacine')
plt.xlabel('Month number')
plt.show()
```

# state and individual vaccinated

```python

keys = [st for st ,df in df.groupby(['State']) ]
plt.bar(keys,df.groupby(['State']).sum()['Total Individuals Vaccinated'])
plt.ylabel('number')
plt.xlabel('state')
plt.xticks(keys,rotation='vertical',size=8)
plt.show()
```

# total individual and second dose in each state

```python
X = [st for st ,df in df.groupby(['State']) ]
Ygirls = df.groupby(['State']).sum()['Total Individuals Vaccinated']
Zboys = df.groupby(['State']).sum()['Second Dose Administered']
  
X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, Ygirls, 0.4, label = 'individual')
plt.bar(X_axis + 0.2, Zboys, 0.4, label = 'second_Dose')
  
plt.xticks(X_axis,X,rotation='vertical', size=8)
plt.xlabel("states")
plt.ylabel("number of dose")
plt.title("vaccine chart")
plt.legend()
plt.show()
```

```python
def histogram_intersection(a, b):
    v = np.minimum(a, b).sum().round(decimals=1)
    return v
```
